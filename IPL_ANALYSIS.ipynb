```python

mat['winner'] = mat['winner'].str.strip().str.title()
mat['toss_winner'] = mat['toss_winner'].str.strip().str.title()
mat['toss_decision'] = mat['toss_decision'].str.strip().str.lower()
mat['venue'] = mat['venue'].str.strip()


deli['batter'] = deli['batter'].str.strip().str.title()
deli['bowler'] = deli['bowler'].str.strip().str.title()
deli['dismissal_kind'] = deli['dismissal_kind'].str.strip().str.lower()
```

```python
mat = mat.dropna(subset=['winner'])
mat['player_of_match'] = mat['player_of_match'].fillna('Unknown')
```

```python
mat.to_csv('matches_cleaned.csv', index=False)
deli.to_csv('deliveries_cleaned.csv', index=False)

print("Saved successfully")
```

```python
import sqlite3

conn = sqlite3.connect(':memory:')

mat.to_sql('matches', conn, index=False, if_exists='replace')
deli.to_sql('deliveries', conn, index=False, if_exists='replace')

print("Tables loaded into SQLite")
```

```python
def run_sql(query):
    return pd.read_sql_query(query, conn)
```

```python
run_sql(""" SELECT winner,
COUNT(*) as total_wins
from matches
GROUP BY winner
ORDER BY total_wins desc
LIMIT 10
""")
```

```python
run_sql( """select ROUND(
    SUM(CASE WHEN toss_winner=winner THEN 1 ELSE 0 END)*100.0/COUNT(*),2
) as toss_win_pct 
FROM matches """)
```

```python
run_sql(""" select toss_decision,count(*) as wins
from matches 
where toss_winner=winner
group by toss_decision """)
```

```python
run_sql(""" select batter,sum(batsman_runs) as total_runs
from deliveries
group by batter
order by total_runs desc
limit 10 """)
```

```python
run_sql(""" select bowler,sum(is_wicket) as total_wicket
from deliveries
where dismissal_kind !='run out'
group by bowler
order by total_wicket desc 
limit 10 """)
```

```python
run_sql(""" select bowler,sum(is_wicket) as wickets
from deliveries
where over between 16 and 20
group by bowler
order by wickets desc
limit 10 """)
```

```python
run_sql(""" select bowler,sum(is_wicket) as wickets
from deliveries
where over between 1 and 6
group by bowler
order by wickets desc
limit 10 """)
```

```python
run_sql(""" select venue,avg(target_runs) as total_runs
from matches
group by venue
order by total_runs desc
limit 5 """)
```

```python
run_sql("""
SELECT m.season, 
       SUM(d.total_runs) as total_runs,
       COUNT(DISTINCT d.match_id) as matches,
       ROUND(SUM(d.total_runs) * 1.0 / COUNT(DISTINCT d.match_id), 2) as avg_runs_per_match
FROM deliveries d
JOIN matches m ON d.match_id = m.id
GROUP BY m.season
ORDER BY m.season
""")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns


wins = mat['winner'].value_counts().head(10)
plt.figure(figsize=(10,6))
sns.barplot(x=wins.values, y=wins.index, palette='Blues_r')
plt.title('Top 10 IPL Teams by Total Wins')
plt.savefig('chart1_team_wins.png', dpi=150)
plt.show()
```

```python


toss_won_match = (mat['toss_winner'] == mat['winner']).sum()
toss_lost_match = (mat['toss_winner'] != mat['winner']).sum()

labels = ['Won Match', 'Lost Match']
sizes = [toss_won_match, toss_lost_match]

plt.figure(figsize=(6,6))
plt.pie(
    sizes,
    labels=labels,
    autopct='%1.1f%%'
)

plt.title('Does Winning the Toss Help?')
plt.show()
```

```python

result = mat.groupby(['season', 'toss_decision']).apply(
    lambda x: (x['toss_winner'] == x['winner']).mean() * 100
).reset_index()
result.columns = ['season', 'toss_decision', 'win_pct']


plt.figure(figsize=(14, 6))
sns.barplot(x='season', y='win_pct', hue='toss_decision', data=result, palette='Set2')
plt.title('Win % by Toss Decision (Bat vs Field) per Season')
plt.xlabel('Season')
plt.ylabel('Win Percentage (%)')
plt.xticks(rotation=45)
plt.legend(title='Toss Decision')
plt.tight_layout()
plt.savefig('chart_bat_vs_field_season.png', dpi=150)
plt.show()
```

```python
top_batsmen = (
    deli.groupby('batter')['batsman_runs']
    .sum()
    .sort_values(ascending=False)
    .head(10)
)

plt.figure(figsize=(10,6))

sns.barplot(
    x=top_batsmen.values,
    y=top_batsmen.index,
    palette='viridis'
)

plt.title('Top 10 IPL Batsmen by Runs')
plt.xlabel('Runs')
plt.ylabel('Batsman')

plt.show()
```

```python

match_runs = deli.groupby('match_id')['total_runs'].sum().reset_index()
match_runs.columns = ['match_id', 'match_total_runs']


match_runs = match_runs.merge(mat[['id', 'season']], left_on='match_id', right_on='id')


avg_season = match_runs.groupby('season')['match_total_runs'].mean().reset_index()


plt.figure(figsize=(12, 5))
plt.plot(avg_season['season'], avg_season['match_total_runs'], marker='o', color='steelblue', linewidth=2)
plt.title('Average Total Runs Per Match by Season')
plt.xlabel('Season')
plt.ylabel('Avg Runs Per Match')
plt.xticks(rotation=45)
plt.grid(axis='y', alpha=0.3)
plt.tight_layout()
plt.savefig('chart_season_runs_trend.png', dpi=150)
plt.show()
```

**Key Business Insights**

**1. Mumbai Indians lead the all-time franchise rankings**
Mumbai Indians have won 144 matches, narrowly ahead of Chennai Super Kings (138) and Kolkata Knight Riders (131). This 6-win gap over a 16-season span shows MI's consistency comes from sustained core-squad retention rather than one standout season — a model other franchises should study for long-term roster planning.

**2. Fielding first after winning the toss is the stronger strategy**
Toss winners who chose to field went on to win 377 matches, compared to only 177 when they chose to bat first — roughly 68% of toss-aligned wins come from chasing. Teams should default to bowling first when conditions are unclear, since knowing the target consistently translates into better execution under pressure.

**3. Virat Kohli and Yuzvendra Chahal are the all-time standout performers**
Kohli leads run-scoring with 8,014 runs, ahead of Dhawan (6,769) and Rohit Sharma (6,630), while Chahal tops the wicket charts with 205 wickets ahead of Chawla (192) and Bravo (183). Franchises retaining such consistent multi-season performers gain a measurable competitive edge in auctions.

**4. Death-over and powerplay specialists are two different skill sets**
In death overs (16-20), Bravo (115 wickets) and Malinga (104) dominate, while in powerplay (overs 1-6), Ishant Sharma (59) and Sandeep Sharma (53) lead — almost no overlap between the two lists. Teams should scout and price these phase-specific skills separately rather than assuming a strong bowler is equally effective across all overs.

**5. Scoring has surged since the Impact Player rule**
Average runs per match rose from around 287-315 in 2009-2017 to a peak of 365.79 in 2024, with 2023 at 350.18 — roughly a 17% jump from the earlier baseline. This shift means bowling units now need to be built around containing higher totals, and franchises should weight death-bowling depth more heavily during auctions.

**6. Certain venues consistently produce high-scoring contests**
The Dr. Y.S. Rajasekhara Reddy ACA-VDCA Stadium recorded the highest average chase target (232.5), followed by the Himachal Pradesh Cricket Association Stadium (203) — both well above the typical venue average. Teams playing fixtures at these grounds should prioritize batting depth and death-over specialists in their playing XI.

**Conclusion**

This analysis of 260,000+ ball-by-ball records across 16 IPL seasons shows that consistency, not single-season brilliance, drives franchise success — Mumbai Indians' 144 wins and CSK's 138 reflect sustained squad stability. Tactically, fielding first is the clear winning strategy, with chasing teams winning 68% of toss-aligned matches, confirming the modern T20 shift toward setting a target. At the player level, Kohli's 8,014 runs and Chahal's 205 wickets highlight rare multi-season excellence, while the near-zero overlap between death-over and powerplay wicket-takers proves bowling talent must be valued by match phase. Most notably, scoring has surged 17% since 2017 (315 to 365.79 avg runs/match by 2024), driven by the Impact Player rule and flatter pitches — meaning franchises must now prioritize death-over bowling depth and phase-specific specialists to stay competitive.
